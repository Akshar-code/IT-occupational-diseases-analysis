```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('data.csv')

# Group the data by Age and Gender, and count the occurrences
age_gender_counts = df.groupby(['Age', 'Gender']).size().unstack(fill_value=0)

# Create a stacked bar chart
ax = age_gender_counts.plot(kind='bar', stacked=True, figsize=(10, 6))

# Customize the chart
plt.title('Distribution of Age and Gender')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Gender', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Add value labels on the bars
for c in ax.containers:
    ax.bar_label(c, label_type='center')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file
df = pd.read_csv('data.csv')

# Define the factors to visualize
factors = [
    'Total computer time last week.',
    'During leisure time last week.',
    '>4 hours without breaks last week.',
    'Mouse use time during last week.',
    'Table height',
    'Chair height',
    'Visibility of the Visual display screen',
    'Navigation through screens',
    'Thermal Comfort at workplace',
    'Noise control at workplace',
    'Illumination'
]

# Define orders for categorical variables
time_order = ['1-9 hours/week', '10-29 hours/week', '≥30 hours/week']
leisure_order = ['<8 hours/week', '8–14 hours/week', '≥15 hours/week']
breaks_order = ['0–1 times/week', '2–4 times/week', '≥5 times/week']
mouse_order = ['0–25%', '50%', '75–100%']
quality_order = ['Bad/Very Bad', 'Regular', 'Good/Excellent']

# Function to create and save charts
def create_factor_chart(df, factor):
    plt.figure(figsize=(12, 8))

    # Apply proper ordering for categorical variables
    if factor == 'Total computer time last week.':
        df[factor] = pd.Categorical(df[factor], categories=time_order, ordered=True)
    elif factor == 'During leisure time last week.':
        df[factor] = pd.Categorical(df[factor], categories=leisure_order, ordered=True)
    elif factor == '>4 hours without breaks last week.':
        df[factor] = pd.Categorical(df[factor], categories=breaks_order, ordered=True)
    elif factor == 'Mouse use time during last week.':
        df[factor] = pd.Categorical(df[factor], categories=mouse_order, ordered=True)
    elif factor in ['Table height', 'Chair height', 'Visibility of the Visual display screen',
                    'Navigation through screens', 'Thermal Comfort at workplace',
                    'Noise control at workplace', 'Illumination']:
        df[factor] = pd.Categorical(df[factor], categories=quality_order, ordered=True)

    # Create horizontal bar plot
    sns.barplot(y='Age', x=factor, hue='Gender', data=df, orient='h')

    plt.title(f'{factor} by Age and Gender')
    plt.ylabel('Age Group')
    plt.xlabel(factor)
    plt.legend(title='Gender', bbox_to_anchor=(1.05, 1), loc='upper left')

    plt.tight_layout()
    plt.savefig(f'{factor.replace(" ", "_").replace(".", "")}.png', dpi=300, bbox_inches='tight')
    plt.close()

# Create charts for all factors
for factor in factors:
    create_factor_chart(df, factor)

print("All charts have been created and saved.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file
df = pd.read_csv('test_data - Sheet2 (1).csv')

# Define the columns we're interested in
columns = [
    'Have your company implemented any ergonomic interventions in the past 12 months?',
    'Do you feel that your organization supports ergonomic practices?',
    'How often do you engage in physical activities outside of work to mitigate the effects of prolonged computer use?'
]

# Function to create and save a chart for each column
def create_chart(column):
    plt.figure(figsize=(12, 6))

    # Create a cross-tabulation of the data
    ct = pd.crosstab([df['Age'], df['Gender']], df[column], normalize='index')

    # Create a stacked bar plot
    ct.plot(kind='bar', stacked=True)

    plt.title(f'{column}\nby Age and Gender')
    plt.xlabel('Age Group and Gender')
    plt.ylabel('Proportion')
    plt.legend(title='Response', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()

    # Save the chart as a PNG file
    filename = f"{column.replace(' ', '_').replace('?', '').replace(',', '')}.png"
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    plt.close()

# Create and save a chart for each column
for column in columns:
    create_chart(column)

print("All charts have been created and saved as individual PNG files.")
```
